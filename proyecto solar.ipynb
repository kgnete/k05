<p style="margin-top:7px; margin-bottom:4px"><FONT style="font-size:22px;" color="#000000"><b>2.- DESCRIPCIÓN TÉCNICA DE LA INSTALACIÓN SOLAR FOTOVOLTAICA</b></font></p>
<a name="REF_HTML:_RC_:2:1"></a>
<p style="margin-top:7px; margin-bottom:4px"><FONT style="font-size:18px;" color="#000000"><b>2.1.- Generalidades</b></font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Todas las instalaciones deberán cumplir con las exigencias de protecciones y seguridad de las personas, y entre ellas las dispuestas en el Reglamento Electrotécnico de Baja Tensión o legislación posterior vigente.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Como principio general, se tiene que asegurar, como mínimo, un grado de aislamiento eléctrico de tipo básico (clase I) para equipos y materiales.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se incluirán todos los elementos necesarios de seguridad para proteger a las personas frente a contactos directos e indirectos.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se recomienda la utilización de equipos y materiales de aislamiento eléctrico de clase II.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se incluirán todas las protecciones necesarias para proteger a la instalación frente a cortocircuitos, sobrecargas y sobretensiones.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Los materiales situados en intemperie se protegerán contra los agentes ambientales, en particular contra el efecto de la radiación solar y la humedad. Todos los equipos expuestos a la intemperie tendrán un grado mínimo de protección IP 65, y los de interior, IP 20.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Los equipos electrónicos de la instalación cumplirán con las directivas comunitarias de seguridad eléctrica y compatibilidad electromagnética.</font></p>

<p style="margin-top:7px; margin-bottom:4px"><FONT style="font-size:18px;" color="#000000"><b>2.3.- Descripción de los equipos</b></font></p>
<a name="REF_HTML:_RC_:2:3:1"></a>
<p style="margin-top:7px; margin-bottom:4px"><FONT color="#000000"><b>2.3.1.- Módulos fotovoltaicos</b></font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se dispone de 8 paneles fotovoltaicos con una potencia nominal de 165 W cada uno. Todos los módulos que integren la instalación serán del mismo modelo, o en el caso de modelos distintos, el diseño debe garantizar totalmente la compatibilidad entre ellos y la ausencia de efectos negativos en la instalación por dicha causa.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Todos los módulos deberán satisfacer las especificaciones UNE-EN 61215 para módulos de silicio cristalino, UNE-EN 61646 para módulos fotovoltaicos de capa delgada, o UNE-EN 62108 para módulos de concentración, así como la especificación UNE-EN 61730-1 y 2 sobre seguridad en módulos FV, Este requisito se justificará mediante la presentación del certificado oficial correspondiente emitido por algún laboratorio acreditado. Incorporarán de forma claramente visible e indeleble el modelo, nombre o logotipo del fabricante, y el número de serie, trazable a la fecha de fabricación, que permita su identificación individual.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se utilizarán módulos que se ajusten a las características técnicas descritas a continuación.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Los módulos deberán llevar los diodos de derivación necesarios para evitar las posibles averías de las células y de sus circuitos por sombreados parciales, que tendrán un grado de protección IP 65.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Para que un módulo resulte aceptable, su potencia máxima y corriente de cortocircuito reales, referidas a condiciones estándar deberán estar comprendidas en el margen del ±5% de los correspondientes valores nominales de catálogo.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Será rechazado cualquier módulo que presente defectos de fabricación, como roturas o manchas en cualquiera de sus elementos, así como falta de alineación en las células, o burbujas en el encapsulante.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Los marcos laterales, si existen, serán de aluminio o de acero inoxidable. Cuando las tensiones nominales en continua sean superiores a 48 V, la estructura del generador y los marcos metálicos de los módulos estarán conectados a una toma de tierra, que será la misma que la del resto de la instalación.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se instalarán los elementos necesarios para la desconexión, de forma independiente y en ambos terminales, de cada una de las ramas del generador.</font></p>
<a name="REF_HTML:_RC_:2:3:1:1"></a>
<p style="margin-top:7px; margin-bottom:4px"><FONT color="#000000"><b><i>2.3.1.1.- Conexionado entre los módulos</i></b></font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Los paneles deben ser escogidos de manera que se permita establecer una configuración modular mediante la interconexión de varias unidades. Mediante asociaciones en serie y en paralelo será posible garantizar la tensión e intensidad requeridas.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Con este fin, la instalación diseñada se compone de 8 ramas con un panel en cada una de ellas.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">De este modo, se consigue una tensión de 30.00 V por rama y una intensidad total de 56.00 A.</font></p>
<a name="REF_HTML:_RC_:2:3:1:2"></a>

<p style="margin-top:7px; margin-bottom:4px"><FONT color="#000000"><b>2.3.2.- Estructura soporte</b></font></p>
<a name="REF_HTML:_RC_:2:3:2:1"></a>
<p style="margin-top:7px; margin-bottom:4px"><FONT color="#000000"><b><i>2.3.2.1.- Aspectos generales</i></b></font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">En todos los casos se dará cumplimiento a lo obligado en el Código Técnico de la Edificación respecto a seguridad.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se dispondrán las estructuras soporte necesarias para montar los módulos incluyendo todos los accesorios y bancadas y/o anclajes, evitando cualquier sombra proyectada sobre los módulos.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">La estructura soporte de los módulos ha de resistir, con los módulos instalados, las sobrecargas de viento y nieve, de acuerdo con lo indicado en el Código Técnico de la Edificación y demás normativa de aplicación. Si está construida con perfiles de acero laminado conformado en frío, cumplirán las normas UNE-EN 10219-1 y UNE-EN 10219-2 para garantizar todas sus características mecánicas y de composición química. Si es de acero galvanizado en caliente, cumplirá las normas UNE-EN ISO 14713 (partes 1, 2 y 3) y UNE-EN ISO 10684 y los espesores cumplirán con los mínimos exigibles en la norma UNE-EN ISO 1461.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">El diseño y la construcción de la estructura y del sistema de fijación de módulos, se realizará para la orientación y el ángulo de inclinación especificado para el generador fotovoltaico, teniendo en cuenta la facilidad de montaje y desmontaje, y la posible necesidad de sustituciones de elementos. Además, permitirá las necesarias dilataciones térmicas sin transmitir cargas que puedan afectar a la integridad de los módulos, siguiendo las indicaciones del fabricante.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">La estructura se protegerá superficialmente contra la acción de los agentes ambientales. La realización de taladros en la estructura se llevará a cabo antes de proceder, en su caso, al galvanizado o protección de la estructura. La tornillería será realizada en acero inoxidable. En el caso de que la estructura sea galvanizada se admitirán tornillos galvanizados, exceptuando la sujeción de los módulos a la misma, que serán de acero inoxidable.</font></p>

<p style="margin-top:7px; margin-bottom:4px"><FONT color="#000000"><b>2.3.3.- Acumuladores</b></font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se recomienda que los acumuladores sean de plomo-ácido, preferentemente estacionarias y de placa tubular. Las baterías estacionarias normalmente se componen de celdas individuales de 2 V, conectadas en serie para alcanzar la tensión deseada (12, 24 o 48 V), aunque también pueden ser utilizadas baterías de estructura compacta (monobloc) para pequeñas instalaciones. No se permitirá el uso de baterías de arranque.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Para asegurar una adecuada recarga de las baterías, la capacidad nominal del acumulador (en Ah) no excederá en 25 veces la corriente (en A) de cortocircuito en CEM del generador fotovoltaico.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">La máxima profundidad de descarga (referida a la capacidad nominal del acumulador) no excederá el 80% en instalaciones donde se prevea que descargas tan profundas no serán frecuentes. En aquellas aplicaciones en las que estas sobredescargas puedan ser habituales, tales como alumbrado público, la máxima profundidad de descarga no superará el 60%.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se protegerá, especialmente frente a sobrecargas, a las baterías con electrolito gelificado, de acuerdo a las recomendaciones del fabricante.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">La capacidad inicial del acumulador será superior al 90% de la capacidad nominal. En cualquier caso, deberán seguirse las recomendaciones del fabricante para aquellas baterías que requieran una carga inicial.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">La autodescarga del acumulador a 20&#176;C no excederá el 6% de su capacidad nominal por mes.</font></p>

<p style="margin-top:7px; margin-bottom:4px"><FONT color="#000000"><b>2.3.7.- Cableado</b></font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Todo el cableado cumplirá con lo establecido en la legislación vigente.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Los conductores necesarios tendrán la sección adecuada para reducir las caídas de tensión y los calentamientos. Concretamente, para cualquier condición de trabajo, los conductores deberán tener la sección suficiente para que la caída de tensión sea inferior, incluyendo cualquier terminal intermedio, al 1,5% de la tensión nominal continua del sistema.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Se incluirá toda la longitud de cables necesaria (parte continua y/o alterna) para cada aplicación concreta, evitando esfuerzos sobre los elementos de la instalación y sobre los propios cables.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Los positivos y negativos de la parte continua de la instalación se conducirán separados, protegidos y señalizados (códigos de colores, etiquetas, etc.) de acuerdo a la normativa vigente.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Los cables de exterior estarán protegidos contra la intemperie.</font></p>
<a name="REF_HTML:_RC_:2:3:8"></a>
<p style="margin-top:7px; margin-bottom:4px"><FONT color="#000000"><b>2.3.8.- Protecciones y puesta a tierra</b></font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">Todas las instalaciones con tensiones nominales superiores a 48 voltios contarán con una toma de tierra a la que estará conectada, como mínimo, la estructura soporte del generador y los marcos metálicos de los módulos.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">El sistema de protecciones asegurará la protección de las personas frente a contactos directos e indirectos. En caso de existir una instalación previa no se alterarán las condiciones de seguridad de la misma.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">La instalación estará protegida frente a cortocircuitos, sobrecargas y sobretensiones. Se prestará especial atención a la protección de la batería frente a cortocircuitos mediante un fusible, disyuntor magnetotérmico u otro elemento que cumpla con esta función.</font></p>

<p style="margin-top:7px; margin-bottom:4px"><FONT color="#000000"><b>2.4.1.- Estimación del consumo de energía</b></font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">La estimación correcta de la energía consumida por el sistema fotovoltaico sólo es sencilla en aquellas aplicaciones en las que se conocen exactamente las características de la carga (por ejemplo, sistemas de telecomunicación). Sin embargo, en otras aplicaciones, como puede ser la electrificación de viviendas, la tarea no resulta fácil pues intervienen multitud de factores que afectan al consumo final de electricidad: tamaño y composición de las familias (edad, formación, etc.), hábitos de los usuarios, capacidad para administrar la energía disponible, etc.</font></p>
<p align="justify" style="margin-top:0px; margin-bottom:10px"><FONT color="#000000">El consumo de las cargas incluirá el servicio de energía eléctrica ofrecido al usuario para distintas aplicaciones (iluminación, TV, frigorífico, bombeo de agua, etc.), e incluirá las pérdidas diarias de energía causadas por el autoconsumo de los equipos (regulador, inversor, etc.).</font></p>
<br>
